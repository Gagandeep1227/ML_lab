```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data          # features (sepal length, sepal width, petal length, petal width)
y_true = iris.target   # actual labels (0,1,2) â€“ not used in clustering, just for comparison

# 2. Create KMeans model
k = 3  # because Iris has 3 species
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

# 3. Fit model and get cluster labels
kmeans.fit(X)
cluster_labels = kmeans.labels_

# 4. View first few results
df = pd.DataFrame(X, columns=iris.feature_names)
df['True_Label'] = y_true
df['Cluster_Label'] = cluster_labels

print(df.head(10))
```

```python
import matplotlib.pyplot as plt

# Use first two features for visualization (sepal length, sepal width)
x1 = X[:, 0]  # sepal length
x2 = X[:, 1]  # sepal width

plt.scatter(x1, x2, c=cluster_labels)
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            marker='X',
            s=200)
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('K-Means Clustering on Iris (k=3)')
plt.show()
```
